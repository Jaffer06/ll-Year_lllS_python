```python
import pandas as pd

# --- Creating dates ---
# Create a single timestamp
date = pd.Timestamp("2025-10-08")
print("Single date:", date)

# Create a date range
date_range = pd.date_range(start="2025-10-01", end="2025-10-10")
print("\nDate Range:")
print(date_range)

# --- Working with current date and time ---
now = pd.Timestamp.now()
print("\nCurrent date and time:", now)
print("Only date:", now.date())
print("Only time:", now.time())

# --- Converting strings to datetime ---
date_str = ["2025-01-01", "2025-05-15", "2025-10-08"]
dates = pd.to_datetime(date_str)
print("\nConverted from string to datetime:")
print(dates)

# --- Extracting parts of dates ---
print("\nYear:", dates[0].year)
print("Month:", dates[0].month)
print("Day:", dates[0].day)

# --- Creating a DataFrame with dates ---
df = pd.DataFrame({
    "Date": pd.date_range("2025-10-01", periods=5, freq="D"),
    "Value": [10, 20, 15, 25, 30]
})
print("\nDataFrame with dates:")
print(df)

# --- Filtering by date ---
filtered = df[df["Date"] > "2025-10-03"]
print("\nFiltered DataFrame (after 2025-10-03):")
print(filtered)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a range of dates
dates = pd.date_range(start="2025-01-01", end="2025-01-15")

# Create some sample values
values = [10, 12, 9, 15, 18, 16, 20, 19, 17, 22, 25, 24, 23, 26, 28]

# Create a DataFrame
df = pd.DataFrame({
    'Date': dates,
    'Value': values
})

# Set 'Date' as the index (helps for time-series plotting)
df.set_index('Date', inplace=True)

# Plot the data
df.plot(title='Values Over Time', marker='*', color='red')

# Add labels and grid
plt.xlabel('Date')
plt.ylabel('Value')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.dates as mdates
import matplotlib.pyplot as plt
import numpy as np
#Generate time-series data
dates = pd.date_range(start="2023-01-01", periods=15, freq="D")
values = np.random.randint(50, 100, size=15)

plt.figure(figsize=(15, 5))
plt.plot(dates, values, marker='o', linestyle='-', color='purple')

plt.title("Time-Series Data")
plt.xlabel("Date")
plt.ylabel("value")

#customize data format
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))  #e.g., Jan
plt.gcf().autofmt_xdate()        #rotate dates
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create random data
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

# Create 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(x, y, z, c='blue', marker='o')

# Labels
ax.set_title("3D Scatter Plot")
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

# 3D data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

#Surface plot
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("3D Surface Plot")
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create two arrays
arr1 = np.linspace(0, 10, 5)
arr2 = np.ones(5)

# Plot the arrays
plt.plot(arr1, label='linspace array', marker='o')
plt.plot(arr2, label='ones array', marker='s')

# Add labels and title
plt.title('Visualization of NumPy Arrays')
plt.xlabel('Index')
plt.ylabel('Value')

# Add legend and grid
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
